```python
#Files to load 
file_to_load = "Resources/missing_grades.csv"


#Read the CSV file into a dataframe
missing_grade_df = pd.read_csv(file_to_load)
missing_grade_df
```

```python
#Get the student CSV file
student_data_to_load = "Resources/students_complete.csv"

#Read the the student data on a table
student_data_df = pd.read_csv(student_data_to_load)

#Add each prefix and suffix to remove to a list 
prefixes_suffixes = ["Dr. ", "Mr. ", "Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]

student_data_df 
```

```python
#iterate through the "prefixes_suffixes" list and replace them with an empty space, "", when it appears in the students name. 
#student_data_df["student_name"].str.replace()

for word in prefixes_suffixes: 
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word, "")

```
